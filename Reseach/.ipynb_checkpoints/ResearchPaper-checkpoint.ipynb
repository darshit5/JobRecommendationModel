```python
import re
from ftfy import fix_text
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.tree import DecisionTreeClassifier
import joblib
import skills_extraction as skills_extraction

# Load dataset:
jd_df = pd.read_csv(r'jd_structured_data.csv')

# Assuming that your dataframe has a 'Sector' column that contains the labels for your data
labels = jd_df['Sector'].values.astype('U')

def ngrams(string, n=3):
    string = fix_text(string)  # fix text
    string = string.encode("ascii", errors="ignore").decode()  # remove non-ascii chars
    string = string.lower()
    chars_to_remove = [")", "(", ".", "|", "[", "]", "{", "}", "'"]
    rx = '[' + re.escape(''.join(chars_to_remove)) + ']'
    string = re.sub(rx, '', string)
    string = string.replace('&', 'and')
    string = string.replace(',', ' ')
    string = string.replace('-', ' ')
    string = string.title()  # normalize case - capital at the start of each word
    string = re.sub(' +', ' ', string).strip()  # get rid of multiple spaces and replace with a single
    string = ' ' + string + ' '  # pad names for ngrams...
    string = re.sub(r'[,-./]|\sBD', r'', string)
    ngrams = zip(*[string[i:] for i in range(n)])
    return [''.join(ngram) for ngram in ngrams]
    
# Split the data into a training set and a testing set
jd_train, jd_test, labels_train, labels_test = train_test_split(jd_df['Processed_JD'].values.astype('U'), labels, test_size=0.2, random_state=42)

# Transform the training and testing sets
vectorizer = TfidfVectorizer(min_df=1, analyzer=ngrams, lowercase=False)
jd_train_transformed = vectorizer.fit_transform(jd_train)
jd_test_transformed = vectorizer.transform(jd_test)

# Save the vectorizer to a file
joblib.dump(vectorizer, 'vectorizer_dt.pkl')

# Train the Decision Tree model on the training set
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(jd_train_transformed, labels_train)

# Save the Decision Tree model to a file
joblib.dump(dt_model, 'dt_model.pkl')

# Load the Decision Tree model from the file
dt_model_loaded = joblib.load('dt_model.pkl')

# Use the loaded model to make predictions on the testing set
labels_pred_dt = dt_model_loaded.predict(jd_test_transformed)

# Build a confusion matrix
print("Decision Tree Confusion Matrix:")
print(confusion_matrix(labels_test, labels_pred_dt))
print("Decision Tree Classification Report:")
print(classification_report(labels_test, labels_pred_dt))
print("Decision Tree Accuracy: ", accuracy_score(labels_test, labels_pred_dt))


# Using the Decision Tree model to predict sector for resume skills
# Load the trained Decision Tree model:
dt_model_loaded = joblib.load('dt_model.pkl')

# Load the vectorizer used during training:
vectorizer_dt = joblib.load('vectorizer_dt.pkl')

# Transform the preprocessed resume skills using the loaded vectorizer:
skills_transformed_dt = vectorizer_dt.transform(preprocessed_skills)

# Predict the sector using Decision Tree:
predicted_sector_dt = dt_model_loaded.predict(skills_transformed_dt)

# Display the predicted sector:
print("Decision Tree Predicted Sector:", predicted_sector_dt[0])

# Filter the jobs in the dataset that belong to the predicted sector.
recommended_jobs_dt = jd_df[jd_df['Sector'] == predicted_sector_dt[0]]

# Display the recommended jobs:
print("Recommended Jobs based on Decision Tree:")
print(recommended_jobs_dt[['Job Title', 'Sector', 'Processed_JD']])
```

```python
import re
from ftfy import fix_text
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import numpy as np
import pandas as pd
import joblib
import skills_extraction as skills_extraction
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Load dataset:
jd_df = pd.read_csv(r'jd_structured_data.csv')

# Assuming that your dataframe has a 'Sector' column that contains the labels for your data
labels = jd_df['Sector'].values.astype('U')

# Split the data into a training set and a testing set
jd_train, jd_test, labels_train, labels_test = train_test_split(jd_df['Processed_JD'].values.astype('U'), labels, test_size=0.2, random_state=42)

# Transform the training and testing sets
vectorizer = TfidfVectorizer(min_df=1, analyzer=ngrams, lowercase=False)
jd_train_transformed = vectorizer.fit_transform(jd_train)
jd_test_transformed = vectorizer.transform(jd_test)

# Save the vectorizer to a file
joblib.dump(vectorizer, 'vectorizer.pkl')

# Define the SVM model
svm = SVC()

# Set the parameters for GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100], 'kernel': ['linear', 'rbf', 'poly', 'sigmoid']}

# Create the GridSearchCV object
grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy')

# Fit the model using GridSearchCV on the training set
grid_search.fit(jd_train_transformed, labels_train)

# Get the best parameters from the grid search
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Use the best model to make predictions on the testing set
best_svm = grid_search.best_estimator_
labels_pred = best_svm.predict(jd_test_transformed)

# Build a confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(labels_test, labels_pred))
print("\nClassification Report:")
print(classification_report(labels_test, labels_pred))
print("\nAccuracy:", accuracy_score(labels_test, labels_pred))

# Save the best model to a file
joblib.dump(best_svm, 'best_svm_model.pkl')

# Load the model from the file
best_svm_loaded = joblib.load('best_svm_model.pkl')

# Load the extracted resume skills:
file_path = r'CV.pdf'
skills = skills_extraction.skills_extractor(file_path)

# Preprocess the resume skills:
preprocessed_skills = [' '.join(ngrams(skill)) for skill in skills]

# Transform the preprocessed resume skills using the loaded vectorizer:
skills_transformed = vectorizer.transform(preprocessed_skills)

# Predict the sector using the best SVM model:
predicted_sector = best_svm_loaded.predict(skills_transformed)

# Display the predicted sector:
print("\nPredicted Sector:", predicted_sector[0])

# Now, you can recommend jobs based on the predicted sector from the best SVM model.
# You may filter the jobs in your dataset that belong to the predicted sector.
recommended_jobs = jd_df[jd_df['Sector'] == predicted_sector[0]]

# Display the recommended jobs:
print("\nRecommended Jobs:")
print(recommended_jobs[['Job Title', 'Sector', 'Processed_JD']])
```
